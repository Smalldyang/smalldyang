由於對於機器學習的模型及語法還不是太熟悉，所以這邊選擇更改課程模型的參數。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("https://raw.githubusercontent.com/ryanchung403/dataset/main/Housing_Dataset_Sample.csv")
df.head()
```

```python
X = df.iloc[:,:5]
y = df['Price']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=54)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
predictions =  reg.predict(X_test)

from sklearn.metrics import r2_score
r2_score(y_test, predictions)
```

中間的語法能改變的參數不多，所以從最後要做回歸曲線的數值來著手，首先我將test_size固定在0.3，也就是以原數據集的30%作為訓練集，再以這訓練集去測試剩下70%的測試集，其實測試得值大部分都落在0.91左右，由此可見改變random_state的值應該不是影響回歸數值的主要原因。
